#**Data Cleaning**

**Filtering Data**

Columns with Only 1 Value

```python
# Check number of unique values for each column
loan_data.nunique(axis = 0)
```

```python
# Drop Columns with only 1 unique value
loan_data.drop(columns = loan_data.columns[loan_data.nunique() == 1], inplace = True)
```

Columns with all unique values

```python
# Drop columns with all unique values
loan_data.drop(columns = loan_data.columns[loan_data.nunique() == len(loan_data)], inplace = True)
```

```python
loan_data.shape
```

Columns with Irrelevant Data

```python
# Drop Columns with irrelevant data
# Drops
# emp_title -> The job title supplied by the Borrower when applying for the loan
# desc -> Loan description provided by the borrower
# title -> The loan title provided by the borrower
# zip_code -> The first 3 numbers of the zip code provided by the borrower in the loan application
# last_pymnt_d -> Last month payment was received
# last_pymnt_amnt -> Last total payment amount received
loan_data.drop(['emp_title', 'desc', 'title', 'zip_code', 'last_pymnt_d', 'last_pymnt_amnt'],axis=1, inplace=True)
```

```python
loan_data.shape
```

```python
loan_data.describe()
```

Drop Rows with Loan status (loan_status variable in the dataset) as Current
since we need to consider only either fully-paid or charged-off loans

```python
loan_data.drop(loan_data[loan_data['loan_status'] == 'Current'].index, inplace=True)
```

```python
loan_data.shape
```

```python
loan_data['loan_status'].value_counts()
```

**Missing Values**

Remove Columns with All Missing Values

```python
100*loan_data.isnull().mean()
```

```python
# Find the columns with all missing values
loan_data.columns[100*loan_data.isnull().mean() == 100]
```

```python
# Drop the columns with 100% i.e. all missing values
loan_data.dropna(axis=1, how='all', inplace=True)
```

Remove Columns with high missing values

```python
# Find columns with high missing values (More than 50%)
loan_data.columns[100*loan_data.isnull().mean() >= 50]
```

```python
# Drop columns with high missing values (More than 50%)
loan_data.drop(['mths_since_last_delinq', 'mths_since_last_record'],axis=1, inplace=True)
```

Missing Value Imputation

```python
# Find if data has missing values for any column
loan_data.columns[loan_data.isnull().any()]
```

```python
loan_data.isnull().sum()
```

```python
# From the above, there are 4 columns with missing values i.e. emp_length, revol_util, last_credit_pull_d and pub_rec_bankruptcies
```

```python
# Remove empty data for last_credit_pull_d since only 2 data points present
loan_data.dropna(axis = 0, subset = ['last_credit_pull_d'], inplace = True)
```

```python
# Remove rows with empty data for revol_util
loan_data.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
```

```python
loan_data.shape
```

```python
loan_data.isnull().sum()
```

```python
# Impute the pub_rec_bankruptcies column
loan_data['pub_rec_bankruptcies'].value_counts()

# Impute the missing values with mode
loan_data['pub_rec_bankruptcies'].fillna(loan_data['pub_rec_bankruptcies'].mode()[0], inplace = True)
loan_data['pub_rec_bankruptcies'].isna().sum()
```

```python
# Impute the emp_length data
loan_data['emp_length'].value_counts()

# Impute the missing values with mode
loan_data['emp_length'].fillna(loan_data['emp_length'].mode()[0], inplace = True)
loan_data['emp_length'].isna().sum()
```

**Standardize data**

Seems we need to standardize data for some of the columns i.e.


*   int_rate
*   emp_length
*   issue_d
*   revol_util



```python
loan_data['int_rate'].unique()
```

```python
# The above values are numeric in nature but because of '%' sign, appears as object type
# Hence, need to convert them to numeric data type
loan_data['int_rate'] = pd.to_numeric(loan_data['int_rate'].apply(lambda x : x.split('%')[0]))
loan_data = loan_data.astype({'int_rate': 'int'})
```

```python
loan_data['int_rate'].dtype
```

```python
loan_data['emp_length'].unique()
```

```python
# Rename the categories >10 years as 10 and <1 years as 0 and remove 'years' word
loan_data['emp_length'] = pd.to_numeric(loan_data['emp_length'].apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

```python
loan_data['emp_length'].unique()
```

```python
import calendar
# Change issue_d to datetime datatype
loan_data['issue_d'].head()

loan_data['issue_d'] = pd.to_datetime(loan_data['issue_d'], format='%b-%y')

# Add some derived data here
loan_data['issue_month'] = loan_data['issue_d'].dt.month.apply(lambda x: calendar.month_abbr[x])
loan_data['issue_year'] = loan_data['issue_d'].dt.year

del loan_data['issue_d']
```

```python
loan_data['revol_util'].unique()
```

```python
# The above values are numeric in nature but because of '%' sign, appears as object type
# Hence, need to convert them to numeric data type
loan_data['revol_util'] = pd.to_numeric(loan_data['revol_util'].apply(lambda x : x.split('%')[0]))
```

```python
loan_data['term'].unique()
```

```python
loan_data.info()
```

```python
loan_data['home_ownership'].unique()
```

```python
# Replace data points having home_ownership as NONE with OTHER since NONE doesn't provide any meaning
loan_data['home_ownership'].replace('NONE', 'OTHER', inplace=True)
```

#**Univariate Analysis**

**Outlier Detection and Removal**\

Segmented Univariate Analysis

```python
# Here we create segments within each numeric variables in order to get insights out of numeric variables
loan_data['loan_amnt_seg'] = pd.cut(loan_data['loan_amnt'], list(range(0, 35001, 5000)),
                                    labels = ['>5000', '5000 - 10000', '10000 - 15000', '15000 - 20000', '20000 - 25000', '25000 - 30000', '30000+'])

loan_data['int_rate_seg'] = pd.cut(loan_data['int_rate'], list(range(0, 26, 5)),
                                   labels = ['0 - 5%', '5% - 10%', '10% - 15%', '15% - 20%', '20%+'])

loan_data = loan_data.astype({'installment': 'int'})
loan_data['installment_seg'] = pd.cut(loan_data['installment'], list(range(0, 1001, 200)),
                                      labels = ['>200', '200 - 400', '400 - 600', '600 - 800', '800+'])

loan_data['annual_inc_seg'] = pd.cut(loan_data['annual_inc'], list(range(0, 150001, 20000)),
                                     labels = ['>20k', '20k - 40k', '40k - 60k', '60k - 80k', '80k - 100k', '100k - 120k', '120k+'])
```

```python
numerical_data = ['loan_amnt_seg', 'int_rate_seg', 'installment_seg', 'annual_inc_seg']
```

```python
for variable in numerical_data:
    viz = sns.countplot(y=variable, data = loan_data[loan_data['loan_status'] == 'Charged Off'])
    viz.set_title(variable.upper().replace("_", " "))
    viz.set_ylabel('Category')
    viz.set_xlabel('Defaulter Count')
    plt.show()
```

4.   Annual income ranging 40k to 60k
